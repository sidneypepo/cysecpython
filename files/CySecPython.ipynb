## O que é engenharia social?
Esta é a técnica mais utilizada por hackers, crackers e criminosos em geral, na qual consiste em um atacante manipular, persuadir ou enganar uma vítima de tal forma que a mesma forneça informações pessoais, credenciais de acesso ou até mesmo acesso lícito ou ilícito de seus dispositivos informáticos sem ter consciência de tal ato.

# **Malwares via spam**

Spam é qualquer tipo de comunicação digital indesejada e não solicitada, muitas vezes um e-mail, que é enviado em massa, trazendo assim um certo desconforto para o usuário que vizualizar a página ou mesmo trazendo informações chamativas para o mesmo, como ofertas ilegítimas, produtos "milagrosos" ou até mesmo apelo sexual. Tudo isso com o principal intuito de fazer o usuário clicar naquele spam. Muitas vezes, spam's são apenas publicidades de algo, mas o que muitos não sabem é que aquela pequena mensagem pode conter um programa malicioso (malware) que, se executado, pode causar grandes problemas para o usuário e seus dispositivos. 

* ##   Spyware:
Este tipo de malware é um programa malicioso que infiltra-se em seu dispositivo e espiona (spy), coleta arquivos, furta informações confidenciais armazenadas no dispositivo da vítima e dados pessoais, como emails, senhas e dados bancários. O spyware, após infectar algo, monitora toda a atividade de aparelhos e faz o envio dos mesmos para o atacante responsável pelo ataque.

* ##   Ransomware:
Como os spywares, este é um tipo de programa malicioso que furta arquivos e dados pessoais dos usuários, mas no caso dos ransomwares, ele criptografa todos os arquivos do dispositivo e exige um pagamento para resgate (ransom) dos arquivos para recuperar o acesso a eles.

Segundo uma das maiores empresas de segurança da informação, a Kaspersky Lab, foi analisado que no panorama mundial das ameaças cibernéticas no primeiro trimestre deste ano, o Brasil está em oitavo lugar no ranking de países que mais recebem spam's diários, como podemos ver no gráfico a seguir:

```
#Importando bibliotecas utilizadas
import pandas as pd
import matplotlib.pyplot as plt

#Importando dados
total_de_spams = pd.read_csv('https://raw.githubusercontent.com/sidneypepo/cysecpython/master/files/dados-spam.csv', usecols=['Total de spams diários'])
 
#Plotando gráfico
plt.style.use('dark_background')
plt.style.context('dark_background')
total_de_spams.plot(color="#ff2929")
 
#Configurações do gráfico
plt.title('Ataques por spam no Brasil da primeira quinzena de Junho\n')
plt.xlabel('Dias')
plt.ylabel('Número de spam\'s (em milhões)')
plt.xlim(1, 15)
plt.ylim(0, 40000000)
plt.grid(True)
plt.show()
 
#Fonte
print("\nFONTE: \n<ESTATÍSTICAS | Mapa em tempo real de ameaças cibernéticas da Kaspersky. \n[S. l.]: Kaspersky Lab, 2020. Disponível em: https://cybermap.kaspersky.com/pt/stats/#country=215&type=kas&period=m. \nAcesso em: 18 jun. 2020.>")
```

## **Roubo de dados via phishing**

Phishing é um termo originado do inglês que se trata de uma maneira de "pescar" (phishing) dados. Essa ação é caracterizada pela tentativa de adquirir ilicitamente informações de um pessoa, sejam senhas, dados financeiros, dados bancários, números de cartões de crédito ou simplesmente informações pessoais. O fraudador utiliza e-mail, aplicativos e sites que são projetados especificamente para roubar dados pessoais.

Segundo a gigante da tecnologia, Trend Micro, o phishing está em ascensão. Segundo dados da empresa, os países que mais sofreram por esse tipo de ataque em 2019, foram os Estados Unidos, Japão e Holanda. Confira no gráfico abaixo a proporção desses ataques:

```
#Importando bibliotecas utilizadas
import numpy as np
import matplotlib.pyplot as plt

#Importando dados
pais_total = [204653, 179869, 38376]
 
#Plotando gráfico
plt.style.use('dark_background')
plt.style.context('dark_background')
fig, ax = plt.subplots()
 
#Configurações do gráfico
ax.set_title('Países com maior número ataques de phishing para roubo de credenciais\n')
paises = ['Holanda', 'EUA', 'Japão']
width = 0.35
ax.bar(paises, pais_total, width, color="#ff2929")
ax.set_ylabel('Total de ataques')
plt.grid(True)
plt.show()
 
#Fonte
print("\nFONTE: \n<The Rising Tide of Credential Phishing: 2.4 Million Attacks Blocked by Trend Micro Cloud App Security in 2019 1H. \n[S. l.]: Trend Micro, 2019. Disponível em: \nhttps://www.trendmicro.com/vinfo/us/security/news/cybercrime-and-digital-threats/the-rising-tide-of-credential-phishing. \nAcesso em: 19 jun. 2020.>")
```

# **Redes Wi-Fi públicas**

Atualmente, ataques cibernéticos não são mais exclusivamente realizados a grande distâncias, mas sim em locais públicos e movimentados, através da rede Wi-Fi local do ambiente. Um bom exemplo desse vetor de ataque é a primeira cena do primeiro episódio da série MR. ROBOT, [em que o protagonista (Elliot) intercepta o tráfego de uma rede e denuncia o dono de um site de pornografia infantil](https://youtu.be/7e0dexhYWls).

Ataques em redes Wi-Fi públicas possuem várias vertentes, que variam da criação de redes Wi-Fi falsas, até interceptação de dados de usuários conectados a esta rede. 

* ##   Man-In-The-Middle: 
O tipo de ataque em rede mais famoso é o Man-In-The-Middle (MITM), em que o atacante "envenena" a rede e a partir disso consegue obter toda a informação que está sendo recebida e enviada pelas vítimas, como demonstrado na figura a seguir: 
<img src="https://www.geekzilla.com.br/wp-content/uploads/2014/10/man-in-the-middle.png" width="400"/>

* ##   Fake AP's:
Este vetor de ataque ocorre de maneira semelhante ao MITM, porém ao invés do atacante apenas se infiltrar em uma rede e "escutar" o trafego da mesma, ele cria uma rede wireless identica a ela e compromete o acesso dos usuários a rede original, dando a ele o acesso total às informações que trafegam na rede falsa.

* ##   Páginas Fake:
Esta forma de ataque não se limita apenas a redes locais, porém é muito mais fácil para o atacante criar uma página fake em uma Wi-Fi local. Seu objetivo é interceptar conexões a determinado site e fazer com que a vítima navegue na página e forneça informações pessoais para o atacante imaginando está no website original.

Atualmente não existem muitas informações prejuízos ou danos causado por estes três ataques internos em redes Wi-Fi, por isso são tão perigosos. As melhores formas de evitar ser vítima de qualquer um deles são:

*   Evitar ao máximo se conectar a redes Wi-Fi abertas/públicas;
*   Sempre utilizar senhas para proteger sua rede wireless residencial;
*   Evitar entrar, navegar e logar em sites que não possuam certificado SSL (cadeado verde fechado ao lado da URL de um site);
*   Evitar fazer compras online através de uma rede Wi-Fi e sempre optar por utilizar as redes móveis de telefonia (LTE/4G/3G);
*   Sempre desconfiar de comportamentos estranhos de todo e qualquer site;
